```python
class User:
  def __init__(self, name, age, gender):
    self.name = name
    self.age = age
    self.gender = gender # Corrected typo from gemder to gender

# インスタンス化する (Instantiate the class)
user = User("侍太郎", "36", "男性") # Moved outside the class definition

print(user.name)
print(user.age)
print(user.gender)
```

```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def printinfo(self):
    print(self.name)
    print(self.age)

# インスタンスを変数に代入 (Instantiate the class)
person = Human("太郎", "20")

person.printinfo()
```
