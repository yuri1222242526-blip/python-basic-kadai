```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(self.name + "は大人です")
        else:
            print(self.name + "は大人ではありません")

humans = []
humans.append(Human("太郎",18))
humans.append(Human("花子", 20))
humans.append(Human("次郎", 25))

for human in humans:
  human.check_adult()
```
